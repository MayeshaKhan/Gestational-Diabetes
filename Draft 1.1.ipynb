Outlier handaling

```python
HDL_std = df3['HDL'].std() #Calculate Standard Deviation
HDL_std
```

```python
std_plus_3 = HDL_Mean + (3 * HDL_std ) 
std_plus_3
```

```python
neg_std_3 = HDL_Mean - (3 * HDL_std )  
neg_std_3
```

```python
df3[df3['HDL'] > std_plus_3]   #Outlier
```

```python
df3[df3['HDL'] < neg_std_3]   #Outlier
```

```python
merge_outliers = df3[(df3['HDL'] > std_plus_3) | (df3['HDL'] < neg_std_3)]   # Merge two type of outliers

merge_outliers
```

```python
merge_outliers.shape
```

```python
preprocessed_df3 = df3[(df3['HDL'] < std_plus_3) & (df3['HDL'] > neg_std_3)]  #drop all outliers
preprocessed_df3
```

```python
sns.histplot(preprocessed_df3['HDL'], kde=True)
```
