```python
import seaborn as sns
columns_with_nulls = ['BMI', 'HDL', 'Sys BP', 'OGTT']
# Box plots to visualize outliers in each feature
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_with_nulls):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df3[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

```python
def identify_outliers_iqr(df, columns):
    outliers = {}
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers[column] = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

outlier_results = identify_outliers_iqr(df3, columns_with_nulls)
for column, outlier_df in outlier_results.items():
    print(f"Outliers in {column}:")
    print(outlier_df)
```

```python
medians = df3[['BMI', 'HDL', 'OGTT']].median()
mean=  df3[['Sys BP']].mean()
# Fill null values with the respective median values
df3['BMI'].fillna(medians['BMI'] , inplace= True)
df3['HDL'].fillna(medians['HDL'], inplace= True)
df3['OGTT'].fillna(medians['OGTT'], inplace= True)
df3['Sys BP'].fillna(mean['Sys BP'], inplace= True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is called 'df' and has columns 'Diabetes' and 'BMI'
sns.boxplot(x='Class Label(GDM /Non GDM)', y='BMI', data=df3)
plt.xlabel("(GDM /Non GDM) (0 = No, 1 = Yes)")
plt.ylabel("BMI")
plt.title("BMI Distribution by Diabetes Status")
plt.show()
```

```python
# Calculate count of individuals for each diabetes category
count_data = df3['Class Label(GDM /Non GDM)'].value_counts().reset_index()
count_data.columns = ['Class Label(GDM /Non GDM)', 'Count']
count_data['Class Label(GDM /Non GDM)'] = count_data['Class Label(GDM /Non GDM)'].map({0: 'Non-GDM', 1: 'GDM'})

# Plotting the count data
sns.barplot(x='Class Label(GDM /Non GDM)', y='Count', data=count_data, palette='pastel')
plt.xlabel("GDM /Non GDM Status")
plt.ylabel("Count")
plt.title("Number of Individuals in GDM and Non-GDM Groups")
plt.show()
```
